```python
import numpy as np

x = np.loadtxt('/home/james/Documents/research/cfd/coanda_auto/output/X.txt')
dldx = np.loadtxt('/home/james/Documents/research/cfd/coanda_auto/output/dLdX.txt')

resx = x[1,:] - x[0,:]
resg = dldx[1,:] - dldx[0,:]

gamma = 1/np.dot(resg,resg)
print(gamma)

xnew = gamma * dldx[1,:]
print(xnew,x[1,:])



def retrieve_lift(folder, case_iter):
    import os
    file = "{}iteration_{}".format(folder,case_iter)
    os.system('rhoSimpleFoam -case {} -postProcess -func forces -latestTime'.format(file))
    forceFold = os.listdir('{}/postProcessing/forces'.format(file))
    forceFold = forceFold[0]
    sel = []
    with open('{}iteration_{}/postProcessing/forces/{}/forces.dat'.format(folder, case_iter, forceFold)) as f:
        data = f.readlines()[3]
    for i in range(len(data)):
        if data[i] == '(' and data[i+1] == '(':
            j = i
            g = 0
            while g < 2:
                if data[j] == ' ':
                    sel.append(j)
                    g = g + 1
                    j = j+1
                else:
                    j = j+1

            break
    lift = float(data[sel[0]:sel[1]])
    return lift
    
```

```python
import numpy as np
import matplotlib.pyplot as plt

N = 260
folder = "/home/james/Documents/research/converged_cases/"
lift = np.zeros(N)
for i in range(N):
    lift[i] = retrieve_lift(folder,i)
```

```python
plt.figure()
plt.plot(lift/lift[0])

t = np.linspace(0,1,N)
plt.figure()
plt.plot(t,x[:N,0]/x[0,0],t,x[:N,1]/x[0,1],t,lift/lift[0])
```
